```python
from bs4 import BeautifulSoup
import requests
```

```python
url1 = "https://www.imdb.com/search/title?title_type=feature&year=1950-01-01,2019-12-31&sort=num_votes,desc&start="
url2 = "&ref_=adv_nxt"
ctr = 1
soupList = []
while ctr < 10000:
    get_url = url1 + str(ctr) + url2
    ctr += 50
    page_source = requests.get(url=get_url, verify=False)
    soup = BeautifulSoup(page_source.text, 'html.parser')
    soupList.append(soup)
```

```python
import string

def filtr(str):
    filtered_string = filter(lambda x: x in string.printable, str)
    return "".join(filtered_string).replace('\n', '').strip()
```

```python
movieParseData = []

for soup in soupList:
    movieList = soup.find_all('div', class_="lister-item mode-advanced")
    for entry in movieList:
        title = "-"
        year = "-"
        runitime = "-"
        genre = "-"
        imdb_rat = "-"
        metascore = "-"
        director = "-"
        gross = "-"
        
        try:
            title = entry.find("h3", class_="lister-item-header").find("a").get_text()
        except:
            pass
        
        try:
            year = entry.find("h3", class_="lister-item-header").find("span", class_="lister-item-year").get_text()
        except:
            pass
        
        try:
            cert = entry.find("p", class_="text-muted").find("span", class_="certificate").get_text()
        except:
            pass
        
        try:
            runtime = entry.find("p", class_="text-muted").find("span", class_="runtime").get_text()
        except:
            pass
        
        try:
            genre = entry.find("p", class_="text-muted").find("span", class_="genre").get_text()
        except:
            pass
        
        try:
            imdb_rat = entry.find("div", class_="ratings-bar").find("div", class_="inline-block ratings-imdb-rating").get_text()
        except:
            pass
        
        try:
            metascore = entry.find("div", class_="ratings-bar").find("div", class_="inline-block ratings-metascore").find("span").get_text()
        except:
            pass
        
        try:
            director = entry.find("p", class_="").find_all("a")[0].get_text()
        except:
            pass
        
        try:
            gross = entry.find("p", class_="sort-num_votes-visible").find_all("span", {"name":"nv"})[1].get_text()
        except:
            pass
        

        data = [filtr(title), filtr(year), filtr(cert), filtr(runtime), filtr(genre), filtr(imdb_rat), filtr(metascore), filtr(director), filtr(gross)]
        movieParseData.append(data)

print(movieParseData)
```

```python
import csv
with open("Team Avion Task 1.csv", "w") as csvFile:
    writer = csv.writer(csvFile)
    writer.writerow(["Title", "Year", "Certificate", "Runtime", "Genre", "IMDb Rating", "Metascore", "Director", "Gross"])
    for data in movieParseData:
        writer.writerow(data)
```
